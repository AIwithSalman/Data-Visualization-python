## Importing libraries and loading the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("Calgary_Business_Licences.csv")


```

```python
print("Dataset shape:", df.shape)
df.head()
```

## Cleaning the data

```python
# Convert date columns to datetime
df['first_iss_dt'] = pd.to_datetime(df['FIRST_ISS_DT'], errors='coerce')
df['exp_dt'] = pd.to_datetime(df['EXP_DT'], errors='coerce')

# Check for missing values in important columns
missing_summary = df[['first_iss_dt', 'exp_dt', 'TRADENAME', 'COMDISTNM', 'LICENCETYPES']].isnull().sum()

print("Missing values summary:")
print(missing_summary)

# Preview cleaned data
df[['TRADENAME', 'COMDISTNM', 'LICENCETYPES', 'first_iss_dt', 'exp_dt']].head()
```

 ## Let's clean up missing values before we build features:
We will drop rows where:

first_iss_dt or exp_dt is missing → required for duration/renewal

COMDISTNM is missing → used for community-based analysis



```python
# Drop rows with missing critical values

df = df.dropna(subset=['first_iss_dt', 'exp_dt', 'COMDISTNM'])

# Confirm shape after dropping
print("New dataset shape after dropping nulls:", df.shape)
```

```python
### we now have a clean dataset with 21,510 records ###
```

Feature Engineering
Now let’s create 3 important columns:

issue_year → the year the license was first issued

license_duration_days → how long the license is valid

renewed → flag if the business is likely renewing the license

```python

# 1. Extract year from issue date

df['issue_year'] = df['first_iss_dt'].dt.year

# 2. Calculate license duration

df['license_duration_days'] = (df['exp_dt'] - df['first_iss_dt']).dt.days

# 3. Create 'renewed' flag (1 if renewal-related status)

df['renewed'] = df['JOBSTATUSDESC'].apply(lambda x: 1 if 'Renewal' in str(x) else 0)

# Preview the result

df[['TRADENAME', 'issue_year', 'license_duration_days', 'JOBSTATUSDESC', 'renewed']].head()

# Why Feature Engineering:
# - To extract the year the license was first issued as 'issue_year'
# - To calculate the license validity duration in days as 'license_duration_days'
# - To create a binary 'renewed' flag indicating if the license status includes a renewal

```

## Exploratory Data Analysis (EDA)

# Importing necessary visualisation libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
licenses_per_year = df['issue_year'].value_counts().sort_index()
sns.lineplot(x=licenses_per_year.index, y=licenses_per_year.values, marker='o')
plt.title('Business Licenses Issued Per Year in Calgary')
plt.xlabel('Year')
plt.ylabel('Number of Licenses')
plt.grid(True)
plt.show()
```

## Top 10 Calgary Communities by Number of Business Licenses

```python
plt.figure(figsize=(12,6))
top_communities = df['COMDISTNM'].value_counts().head(10)
sns.barplot(x=top_communities.values, y=top_communities.index, palette='viridis')
plt.title('Top 10 Calgary Communities by Number of Business Licenses')
plt.xlabel('Number of Licenses')
plt.ylabel('Community Name')
plt.show()
```

## Insights:  Beltline is one of Calgary’s busiest business communities

# Top License Types

```python
plt.figure(figsize=(12,6))
top_license_types = df['LICENCETYPES'].value_counts().head(10)
sns.barplot(x=top_license_types.values, y=top_license_types.index, palette='magma')
plt.title('Top 10 Business License Types in Calgary')
plt.xlabel('Number of Licenses')
plt.ylabel('License Type')
plt.show()
```

# License status (JOBSTATUSDESC) across the top 10 communities with most licenses.

```python
# Filter top 10 communities
top_10_communities = df['COMDISTNM'].value_counts().head(10).index
df_top_communities = df[df['COMDISTNM'].isin(top_10_communities)]

# Aggregate counts by community and job status
status_distribution = df_top_communities.groupby(['COMDISTNM', 'JOBSTATUSDESC']).size().reset_index(name='count')

# Pivot for plotting
status_pivot = status_distribution.pivot(index='COMDISTNM', columns='JOBSTATUSDESC', values='count').fillna(0)

# Plot stacked bar chart
status_pivot.plot(kind='bar', stacked=True, figsize=(14,7), colormap='tab20')
plt.title('Business License Status Distribution Across Top 10 Calgary Communities')
plt.xlabel('Community Name')
plt.ylabel('Number of Licenses')
plt.xticks(rotation=45)
plt.legend(title='License Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

License Renewal Rate Over Years

```python
# Calculate renewal counts and total counts per year
renewal_summary = df.groupby('issue_year')['renewed'].agg(['sum', 'count'])
renewal_summary['renewal_rate'] = renewal_summary['sum'] / renewal_summary['count']

# Plot renewal rate over years
plt.figure(figsize=(12,6))
sns.lineplot(x=renewal_summary.index, y=renewal_summary['renewal_rate'], marker='o')
plt.title('Business License Renewal Rate Over Years')
plt.xlabel('Issue Year')
plt.ylabel('Renewal Rate')
plt.ylim(0,1)
plt.grid(True)

# Add data labels
for x, y in zip(renewal_summary.index, renewal_summary['renewal_rate']):
    plt.text(x, y + 0.02, f"{y:.2f}", ha='center', fontsize=9)

plt.show()
```

```python
# Calculate renewal rate by community
renewal_by_community = df.groupby('COMDISTNM')['renewed'].agg(['sum', 'count'])
renewal_by_community['renewal_rate'] = renewal_by_community['sum'] / renewal_by_community['count']

# Sort to get top 10 communities by renewal rate
top_10_renewal = renewal_by_community.sort_values('renewal_rate', ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_10_renewal['renewal_rate'], y=top_10_renewal.index, palette='coolwarm')
plt.title('Top 10 Calgary Communities by Business License Renewal Rate')
plt.xlabel('Renewal Rate')
plt.ylabel('Community Name')
plt.xlim(0,1)
plt.show()
```

```python
# Create duration buckets in days (adjust as needed)
bins = [0, 365, 1095, 1825, df['license_duration_days'].max()]
labels = ['<1 year', '1-3 years', '3-5 years', '5+ years']

df['duration_bucket'] = pd.cut(df['license_duration_days'], bins=bins, labels=labels, include_lowest=True)

# Group by duration bucket and calculate renewal rate
duration_renewal = df.groupby('duration_bucket')['renewed'].agg(['sum', 'count'])
duration_renewal['renewal_rate'] = duration_renewal['sum'] / duration_renewal['count']

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=duration_renewal.index, y=duration_renewal['renewal_rate'], palette='Blues_d')
plt.title('Business License Renewal Rate by License Duration')
plt.xlabel('License Duration')
plt.ylabel('Renewal Rate')
plt.ylim(0, 1)
plt.show()
```

## Interpretation: License Duration vs Renewal Rate

- Businesses holding licenses for **longer durations** (over 1 year) show **higher renewal rates**, indicating more stable and committed operations.
- Shorter license durations have lower renewal rates, possibly representing temporary or seasonal businesses.
- Encouraging longer license terms could help improve overall business retention and economic stability in Calgary.
